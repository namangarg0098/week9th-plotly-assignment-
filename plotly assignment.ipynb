```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.
```

```python
tips = plt.data.tips()
tips.head()
```

```python
fig = plt.histogram(tips, x='sex', y='total_bill', pattern_shape='smoker', color='day')
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
Plotly Express has a function called Distplot that creates a histogram and kernel density estimation plot. It shows us how a continuous variable is distributed in a dataset.

Displot function is depricated in plotly for function histogram.
```

```python
import pandas as pd

df = pd.DataFrame({'numbers': [1, 1, 1, 2, 2, 3, 3, 4, 5, 5]})


fig = plt.histogram(df, x="numbers", marginal="rug", histnorm="probability density")
fig.show()
```
